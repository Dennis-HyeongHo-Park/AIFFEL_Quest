#손글씨 분류

필요한 모델 페키지 및 라이브러리 불러오기

```python
import pandas as pd
import numpy as np


from sklearn.datasets import load_digits # 손글씨 데이터 셋 불러오기
from sklearn.model_selection import train_test_split # 데이터를 train과 test 데이터셋으로 나눠주는 함수

from sklearn.tree import DecisionTreeClassifier   # 결정트리 모델 패키지
from sklearn.ensemble import RandomForestClassifier  # 랜덤 포레스트 모델 패키지
from sklearn import svm   # 서포트 백터 모델 패키지
from sklearn.linear_model import SGDClassifier  # 확률 경사하강법 모델 피키지
from sklearn.linear_model import LogisticRegression # 선형 회귀 모델 피키지

from sklearn.metrics import classification_report # 정답 결과 확인
```

데이터 준비

```python

digits = load_digits()
digits.keys()
```

데이터 이해하기.

```python
# Featur Data 지정하기
digits_data = digits.data
digits_data.shape
```

```python
pd.DataFrame(digits_data)
```

```python
# Label Data 지정하기
digits_label = digits.target
digits_label.shape
```

```python
# Target Names 출력해보기
digits_label[:20]
```

```python
digit_names = digits.target_names
digit_names
```

```python
# 데이터 Describe해 보기
print(digits.DESCR)
```

train, test 데이터 분리하기

```python
# train과 test 데이터 분리하기.
X_train, X_test, y_train, y_test = train_test_split(digits_data,
                                                    digits_label,
                                                    test_size = 0.2,
                                                    random_state = 7)
```

```python
print( 'X_train 개수 :', len(X_train), 'X_test 개수 :', len(X_test))
```

###다양한 모델로 학습시켜보기

#### Decision Tree 사용해 보기

```python
decision_tree = DecisionTreeClassifier(random_state = 32) # 결정트리 모델 객체 만들기(풀기).

print(decision_tree._estimator_type) # 모델 타입 확인해보기
```

```python
# 결정트리로 학습시키기
decision_tree.fit(X_train, y_train)
```

```python
# 학습후 예측하기
y_pred = decision_tree.predict(X_test)
y_pred
```

```python
# 에측된 결과 와 정답지 비교하기
print(classification_report(y_test, y_pred))
```

#### Random Forest 사용해보기

```python
# 모델 페키지 객체 설정하기
random_forest = RandomForestClassifier(random_state = 32)
```

```python
# 모델 타입 보기
print(random_forest._estimator_type)
```

```python
# 모델 학습시기키
random_forest.fit(X_train, y_train)
```

```python
# 학습한 모델로 예측해보기
y_pred = random_forest.predict(X_test)
y_pred
```

```python
# 예측한 결과 정답지와 비교하기
print(classification_report(y_test, y_pred))
```

```python
# 모델 객체 생성하기
svm_model = svm.SVC()

# 생성된 모델 타입 확인하기
print(svm_model._estimator_type)
```

### 6.  모델 평가해보기

우선 Decision tree 경우 모든 평가지표가 100오 예측했다. 이게 과적합인가?? 아니면 학습을 잘못된 방법으로 했나 싶다. 아니면 . 완벽하게 학습한건가... 싶다 아직 이런 부분에서 해석하기 힘들다.
만야 Decision tree모델에 이상이 없다면 가장 완벽할 것이다.  
Decision tree을 제외하고 나머지를 본다면,  음성을 예측하고 실제 음성인것, 양성을 예측하고 실제 양성인 경우는 SVM과 Logistic Regression의 성이니 다음으로 좋은것 같다 그중 Logistic Regression 이지 않을까 생각한다.
이유는 암의 판단의 경우 양성을 예상했지만 음성인 결과보다 음성을 예상했지만 양성인 경우를 더 조심해야 한다고 배웠다. 그래서 Logistic Regression의 경우는 그나마 85% 로 다른 모델에 비해 나은 성능을 보인것이 아닌게 생각된다.
